```python
#Generates 3D GC-MS total ion count chromatograph
# Oct 2019 Will Cheney

import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib.collections import PolyCollection

```

```python
timeMS = [[],[],[],[]]   #List/Container for x time points
abundance = [[],[],[],[],[],[],[],[]] # List of lists for y abundance values from different data sets
rows = 5673 # Number of data points


#Open first CSV file by giving it a file path. Reader object reads rows from the CSV
#There are 3 rows before the data the first for loop skips these first three rows


with open('/Users/willcheney/GC-MS sterols/2020-01-31 For thesis/Cholesterol std.CSV',mode = 'rt',encoding= 'UTF-8')as csvFile:
    reader = csv.reader(csvFile)
    Peaks = [(0,0),(0,0)]
    max0=0
    reset = False
    for i in range(3):
        next(reader)

    for row in reader: #reads specified number of row
        
        if float(row[0])>26:
            break
        abundance[0].append(float(row[1]))# add Abundance values to Y list


        timeMS[0].append(float(row[0])) #add time values to X list

        

with open('/Users/willcheney/GC-MS sterols/2020-01-31 For thesis/D2 chroma.CSV',mode = 'rt',encoding= 'UTF-8')as csvFile:
    reader = csv.reader(csvFile)
    Peaks = [(0,0),(0,0)]
    max0=0
    reset = False
    for i in range(3):
        next(reader)

    for row in reader: #reads specified number of row
        
        if float(row[0])>26:
            break
        abundance[1].append(float(row[1]))# add Abundance values to Y list


        timeMS[1].append(float(row[0])) #add time values to X list

with open('/Users/willcheney/GC-MS sterols/2020-01-31 For thesis/WCY71.CSV',mode = 'rt',encoding= 'UTF-8')as csvFile:
    reader = csv.reader(csvFile)
    Peaks = [(0,0),(0,0)]
    max0=0
    reset = False
    for i in range(3):
        next(reader)

    for row in reader: #reads specified number of row
        if float(row[0])>26:
            break
        abundance[2].append(float(row[1]))# add Abundance values to Y list


        timeMS[2].append(float(row[0])) #add time values to X list

        
with open('/Users/willcheney/GC-MS sterols/2020-01-31 For thesis/ergo std 100.CSV',mode = 'rt',encoding= 'UTF-8')as csvFile:
    reader = csv.reader(csvFile)
    Peaks = [(0,0),(0,0)]
    max0=0
    reset = False
    for i in range(3):
        next(reader)

    for row in reader: #reads specified number of row
        
        if float(row[0])>26:
            break
        abundance[3].append(float(row[1]))# add Abundance values to Y list


        timeMS[3].append(float(row[0])) #add time values to X list
        
        
```

```python


fig = plt.figure()
ax = plt.axes(projection='3d')

#Set time limits
lowerbound = 10 
upperbound = 16

for i in range(len(timeMS)):
    
    bounded_values = [j for j in timeMS[i] if (j > lowerbound and j < upperbound)]
    abundance[i] = abundance[i][timeMS[i].index((bounded_values[0])): (timeMS[i].index((bounded_values[-1])) + 1)]
    timeMS[i] = bounded_values

ax.plot3D(timeMS[0] , [1] * len(timeMS[0]), abundance[0],'#e84118',linewidth= 1, zorder = 5)
ax.plot3D(timeMS[3] ,[2] * len(timeMS[3]), abundance[3],'#353b48',linewidth= 1, zorder = 4)

ax.plot3D(timeMS[1] ,[3] * len(timeMS[1]) ,abundance[1],'#00a8ff',linewidth= 1, zorder = 3)
ax.plot3D(timeMS[2],[4] * len(timeMS[2]),abundance[2],'#8c7ae6',linewidth= 1, zorder = 2)

yline = np.linspace(0.5, 4.5,1000)

ax.plot3D([timeMS[0][abundance[0].index(max(abundance[0]))]] * len(yline), yline, [0] * len(yline), linestyle = 'dashed', color = '#e84118', zorder = 0, alpha = 0.75 )
#ax.scatter(timeMS[2][abundance[2].index(max(abundance[2]))], 0.5, [0], color = '#8c7ae6', zorder = 0, alpha = 0.75 )
#ax.scatter(timeMS[1][abundance[1].index(max(abundance[1]))], 0.5, [0], color = '#00a8ff', zorder = 0, alpha = 0.75 )

#ax.set_yticks([1,2,10])
ax.set_yticks([])
ax.set_zticks([0, (2*10**6),(4*10**6), (6*10**6), (8*10**6)])
ax.set_zticklabels([])
fig.tight_layout()
plt.xlabel('Time (Min)')
ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
#ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.xaxis.grid()
fig = plt.gcf()
plt.show()
```
